```python
import pandas as pd
import numpy as np
```

```python
student_df = pd.read_csv("student_info.csv")
tuition_df = pd.read_csv("tuition_fees.csv")
payment_df = pd.read_csv("payment_records.csv")

# Kiểm tra nhanh
print(student_df.head())
print(tuition_df.head())
print(payment_df.head())
```

```python
student_df['student_id'] = student_df['student_id'].astype(str).str.strip()

student_df['full_name'] = (
    student_df['full_name']
    .astype(str)
    .str.strip()
    .str.title()
)

student_df['program_type'] = (
    student_df['program_type']
    .astype(str)
    .str.strip()
    .str.lower()
)
```

```python
def normalize_semester(x):
    if pd.isna(x):
        return np.nan
    
    x = str(x)
    year = ''.join(filter(str.isdigit, x))[:4]
    
    if '1' in x:
        return f"{year}-1"
    elif '2' in x:
        return f"{year}-2"
    else:
        return np.nan
```

```python
tuition_df['semester'] = tuition_df['semester'].apply(normalize_semester)
payment_df['semester'] = payment_df['semester'].apply(normalize_semester)
```

```python
def clean_currency(x):
    if pd.isna(x):
        return 0.0
    
    x = str(x)
    x = x.replace(',', '').replace('$', '').strip()
    
    try:
        value = float(x)
        if value < 0:
            return 0.0
        return value
    except:
        return 0.0
```

```python
tuition_df['tuition_amount'] = tuition_df['tuition_amount'].apply(clean_currency)
payment_df['amount_paid'] = payment_df['amount_paid'].apply(clean_currency)
```

```python
tuition_df['due_date'] = pd.to_datetime(
    tuition_df['due_date'], errors='coerce'
)

payment_df['payment_date'] = pd.to_datetime(
    payment_df['payment_date'], errors='coerce'
)
```
