```python
import pandas as pd

df = pd.read_csv("../final_report/kết quả phân tích/tuition_payment_merged.csv", parse_dates=["due_date", "payment_date"])
```

```python
df["late_payment"] = df["payment_date"] > df["due_date"]
```

```python
pivot_late_rate = pd.pivot_table(
    df,
    values="late_payment",
    index="semester",
    columns="program_type",
    aggfunc="mean"
)
pivot_late_rate
```

```python
pivot_amount = pd.pivot_table(
    df,
    values=["tuition_amount", "amount_paid"],
    index=["semester", "program_type"],
    aggfunc="sum"
)
pivot_amount
```

```python
pivot_unstack = pivot_amount.unstack()

pivot_stack = pivot_unstack.stack()

pivot_stack
```
