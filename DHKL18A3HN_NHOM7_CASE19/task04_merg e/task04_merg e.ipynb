```python
import pandas as pd

student_info = pd.read_csv("../data_clean/student_info_clean.csv")
tuition_fees = pd.read_csv("../data_clean/tuition_fees_clean.csv")

tuition_with_program = tuition_fees.merge(
    student_info[["student_id", "program_type"]],
    on="student_id",
    how="left"
)
```

```python
payment_records = pd.read_csv("../data_clean/payment_records_clean.csv")

merged_data = tuition_with_program.merge(
    payment_records,
    on=["student_id", "semester"],
    how="left"
)
```

```python
unpaid_tuition = merged_data[
    merged_data["amount_paid"].isna()
]

unpaid_tuition.head()
```

```python
invalid_payments = payment_records.merge(
    tuition_fees,
    on=["student_id", "semester"],
    how="left",
    indicator=True
)

invalid_payments = invalid_payments[
    invalid_payments["_merge"] == "left_only"
]

invalid_payments.head()
```

```python
merged_data.to_csv(
    "../final_report/kết quả phân tích/tuition_payment_merged.csv",
    index=False
)
```
